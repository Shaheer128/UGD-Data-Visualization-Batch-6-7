```python
new_arr1=np.random.randint(low=0,high=100,size=20).reshape(4,5)
```

```python
print(new_arr1)
```

```python
print(new_arr1[2][0])
```

```python
new_arr1>50
```

```python
#retrieving all odd number that are greater than 50
new_arr1[(new_arr1 % 2 !=0)& (new_arr1>50)]
```

```python
#retrieving all even number that are greater than 50
new_arr1[(new_arr1>50)&(new_arr1 % 2 ==0)]
new_arr1
```

```python
#replacing with zeroes
new_arr1[(new_arr1 % 2 !=0)& (new_arr1>50)]=0
new_arr1
```

```python
x=np.linspace(-30,30,100)
```

```python
y=np.sin(x)
```

```python
y1=np.cos(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("COS Waves")
plt.xlabel("X Values")
plt.ylabel("COS values")
plt.plot(x,y1,color='red')
plt.show()
```

```python
y=x**2+2*x+4
```

```python
plt.title("Y Waves")
plt.xlabel("X Values")
plt.ylabel("Y values")
plt.plot(x,y,color='red')
plt.show()
```
